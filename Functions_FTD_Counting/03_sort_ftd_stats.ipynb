```python
## Load pickle for exports and imports of data  
import pickle 
def load_obj(path):
    with open(path, 'rb') as f:
        return pickle.load(f)
    
def save_obj(obj, path ):
    with open(path, 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)
```

```python
import_path = '../Resources/symbol_success_list_new.pkl'
datadict_success_list = load_obj(import_path)

len(datadict_success_list)
```

```python
dataframe = pd.DataFrame()

for symbol in datadict_success_list:
    import_path = '../FilesExport_Complete_DataDicts_2/'+symbol+'_updated_data_dict.pkl'
    import_data = load_obj(import_path)
    
    data = import_data[symbol]['ftd_stats']
    data['isEtf'] = import_data[symbol]['companyProfile']['isEtf'] 

    empty_dict = {}
    empty_dict[symbol] = data
       
    if symbol == 0:
        dataframe = pd.DataFrame.from_dict(empty_dict,orient='index')
    else: 
        dataframe2 = pd.DataFrame.from_dict(empty_dict,orient='index')
        dataframe = dataframe.append(dataframe2)
        

## Keep only FTD related columns (drop volume data)
dataframe = dataframe[['total_ftds_all','total_ftds_2021','total_ftds_ytd','total_ftds_month',
                      'os_ftd_pct_all','os_ftd_pct_2021','os_ftd_pct_ytd','os_ftd_pct_month',
                      'float_ftd_pct_all','float_ftd_pct_2021','float_ftd_pct_ytd','float_ftd_pct_month',
                      'isEtf']]
        
len(dataframe)
```
