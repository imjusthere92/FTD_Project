```python
## data = data_dict[symbol]
## use like 
## data_dict[symbol] = do_ftd_calculations(data_dict[symbol])

def do_ftd_calculations(data):
    df = data['dataFrame']
    total_ftds_all = df.QUANTITY_FAILS.sum() ## Add to data['ftd_stats']
    df2 = df.loc['2021':]
    total_ftds_2021 = df2.QUANTITY_FAILS.sum() ## Add to data['ftd_stats'] 
    ftd_pct_all = total_ftds_2021 / total_ftds_all * 100 ## Add to data['ftd_stats']
    outstandingShares = data['Fundamentals']['sharesOutstanding']
    ftd_pct_2021 = total_ftds_2021 / outstandingShares * 100 ## Add to data['ftd_stats']
    
    ## All data 
    totalvolume1 = short_df.ShortVolumeNSDQ.sum()
    totalvolume2 = short_df.ShortVolumeNYSE.sum()
    totalshortvolume = totalvolume1+totalvolume2

    totalvolume1 = short_df.ShortExemptVolumeNSDQ.sum()
    totalvolume2 = short_df.ShortExemptVolumeNYSE.sum()
    totalshortexemptvolume = totalvolume1+totalvolume2

    totalvolume1 = short_df.TotalVolumeNSDQ.sum()
    totalvolume2 = short_df.TotalVolumeNYSE.sum()
    total_volume_nsdq_data= totalvolume1+totalvolume2

    total_volume_fmp_data = short_df.volume.sum()

    short_volume_FMPpct_all_data = totalshortvolume / total_volume_fmp_data * 100 ## Add to data['ftd_stats']
    short_volume_NSDQpct_all_data= totalshortvolume / total_volume_nsdq_data * 100 ## Add to data['ftd_stats']

    shortexempt_volume_FMPpct_all_data = totalshortexemptvolume / total_volume_fmp_data * 100 ## Add to data['ftd_stats']
    shortexempt_volume_NSDQpct_all_data= totalshortexemptvolume / total_volume_nsdq_data * 100 ## Add to data['ftd_stats']
    
    ## 2021 data
    totalvolume1 = df2.ShortVolumeNSDQ.sum()
    totalvolume2 = df2.ShortVolumeNYSE.sum()
    totalshortvolume = totalvolume1+totalvolume2
    totalshortvolume2021 = totalshortvolume

    totalvolume1 = df2.ShortExemptVolumeNSDQ.sum()
    totalvolume2 = df2.ShortExemptVolumeNYSE.sum()
    totalshortexemptvolume = totalvolume1+totalvolume2
    totalshortexemptvolume2021 = totalshortexemptvolume

    totalvolume1 = df2.TotalVolumeNSDQ.sum()
    totalvolume2 = df2.TotalVolumeNYSE.sum()
    total_volume_nsdq_data= totalvolume1+totalvolume2

    total_volume_fmp_data = df2.volume.sum()

    short_volume_FMPpct_2021 = totalshortvolume / total_volume_fmp_data * 100 ## Add to data['ftd_stats']
    short_volume_NSDQpct_2021= totalshortvolume / total_volume_nsdq_data * 100 ## Add to data['ftd_stats']

    shortexempt_volume_FMPpct_2021 = totalshortexemptvolume / total_volume_fmp_data * 100 ## Add to data['ftd_stats']
    shortexempt_volume_NSDQpct_2021= totalshortexemptvolume / total_volume_nsdq_data * 100 ## Add to data['ftd_stats']
    
    ftd_stats = {
        'total_ftds_all':total_ftds_all,
        'total_ftds_2021':total_ftds_2021,
        'ftd_pct_all':ftd_pct_all,
        'ftd_pct_2021':ftd_pct_2021,
        'short_volume_NSDQpct_all_data':short_volume_NSDQpct_all_data,
        'shortexempt_volume_NSDQpct_all_data':shortexempt_volume_NSDQpct_all_data,
        'short_volume_NSDQpct_2021':short_volume_NSDQpct_2021,
        'shortexempt_volume_NSDQpct_2021':shortexempt_volume_NSDQpct_2021,
        'short_volume_FMPpct_all_data':short_volume_FMPpct_all_data,
        'shortexempt_volume_FMPpct_all_data':shortexempt_volume_FMPpct_all_data,
        'short_volume_FMPpct_2021':short_volume_FMPpct_2021,
        'shortexempt_volume_FMPpct_2021':shortexempt_volume_FMPpct_2021
    }
    
    ## Add to data passed to function, then return 
    data['ftd_stats'] = ftd_stats
    return data 
```

```python
data_dict['AAPL'] = do_ftd_calculations(data_dict['AAPL'])
data_dict['AAPL']
```
