```python
## Use pickle module to import and export and save files
import pickle
def load_obj(path):
    with open(path, 'rb') as f:
        return pickle.load(f)
def save_obj(obj, path ):
    with open(path, 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)
```

```python
## Create a symbol_list from symbol_df
symbol_list = []
for i in symbol_df['SYMBOL']:
    symbol_list.append(i)
    
len(symbol_list)
```

```python
## Read full_key_list
path = Path('../Resources/read_04_full_key_list.pkl')
full_key_list = load_obj(path)
```

```python
#full_key_list
```

```python
counter = 0  ## Stops this for loop from printing a million things 
for i in full_key_list:
    if counter <= 5:
        print(i)
    counter +=1
```

```python
## Remove all symbols from full_key_list from symbol_list, then use symbol_list
for i in full_key_list:
    symbol_list.remove(i)
len(symbol_list)
```

```python
failed_symbol_list = []
data_dict = {}
for i in symbol_list:
    try:
        symbol = str(i)
    except TypeError:
        continue
    try:
        path = '../FilesExportIndividualStockDFs/'+symbol+'_combined_df.csv'
        data = pd.read_csv(path)
        data_dict[symbol] = data
    except FileNotFoundError:
        failed_symbol_list.append(symbol)
        continue
    
```

```python
len(data_dict)
```

```python
## Export leftover data and continue in new notebook file 
path = Path('../Resources/leftover_data_dict.pkl')
save_obj(data_dict,path)
```
