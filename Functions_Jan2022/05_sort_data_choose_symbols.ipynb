```python
## Load pickle for exports and imports of data  
import pickle 
def load_obj(path):
    with open(path, 'rb') as f:
        return pickle.load(f)
    
def save_obj(obj, path ):
    with open(path, 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)
```

```python
## Calculations are already complete inside ftd_stats
## So to find stocks with highest FTDs overall... 

greatest_os_ftd_pct_all = 0 
greatest_os_ftd_pct_2021= 0 
greatest_os_ftd_pct_ytd = 0
greatest_float_ftd_pct_all = 0
greatest_float_ftd_pct_2021 = 0
greatest_float_ftd_pct_ytd = 0

for symbol in datadict_success_list:
    import_path = '../FilesExport_Complete_DataDicts/'+symbol+'_data_dict.pkl'    
    data = load_obj(import_path)
    
#     ## Shares outstanding
#     sharesOutstanding = data[symbol]['Fundamentals']['sharesOutstanding']
#     ## Float shares
#     data[symbol]['Fundamentals']['floatShares']
    
    ## os_ftd_pct_all - all time count of FTDs for the dataset, relative to outstanding shares 
    os_ftd_pct_all = data[symbol]['ftd_stats']['os_ftd_pct_all']
    ## os_ftd_pct_2021 - pct of FTDs relative to outstanding shares, for Jan1st to Dec31st 2021
    os_ftd_pct_2021 = data[symbol]['ftd_stats']['os_ftd_pct_2021']
    ## os_ftd_pct_ytd - pct of FTDs relative to outstanding shares, for Dec1st 2020 to Dec31st 2021
    os_ftd_pct_ytd = data[symbol]['ftd_stats']['os_ftd_pct_ytd']
    
    
## Note - ETFs do not have data for Float. Float can only be calculated with Stocks, while 
##        outstandingShares can be calculated with either Stocks or ETFs. 
    
    ## float_ftd_pct_all - all time count of FTDs for the dataset, relative to Float shares 
    float_ftd_pct_all = data[symbol]['ftd_stats']['float_ftd_pct_all']
    ## float_ftd_pct_2021 - pct of FTDs relative to Float shares, for Jan1st to Dec31st 2021
    float_ftd_pct_2021 = data[symbol]['ftd_stats']['float_ftd_pct_2021']
    ## float_ftd_pct_ytd - pct of FTDs relative to Float shares, for Dec1st 2020 to Dec31st 2021
    float_ftd_pct_ytd = data[symbol]['ftd_stats']['float_ftd_pct_ytd'] 
    
    if symbol == 0:
        greatest_os_ftd_pct_all = os_ftd_pct_all
        greatest_os_ftd_pct_2021 = os_ftd_pct_2021
        greatest_os_ftd_pct_ytd = os_ftd_pct_ytd
        
        greatest_float_ftd_pct_all = float_ftd_pct_all
        greatest_float_ftd_pct_2021 = float_ftd_pct_2021
        greatest_float_ftd_pct_ytd = float_ftd_pct_ytd
        
    ## Outstanding Shares     
    elif os_ftd_pct_all > greatest_os_ftd_pct_all:
        greatest_os_ftd_pct_all = os_ftd_pct_all
        greatest_os_ftd_pct_all_symbol = symbol
        
    elif os_ftd_pct_2021 > greatest_os_ftd_pct_2021:
        greatest_os_ftd_pct_2021 = os_ftd_pct_2021
        greatest_os_ftd_pct_2021_symbol = symbol 
        
    elif os_ftd_pct_ytd > greatest_os_ftd_pct_ytd:
        greatest_os_ftd_pct_ytd = os_ftd_pct_ytd
        greatest_os_ftd_pct_ytd_symbol = symbol 
        
    ## Float Shares    
    elif float_ftd_pct_all > greatest_float_ftd_pct_all:
        greatest_float_ftd_pct_all = float_ftd_pct_all
        greatest_float_ftd_pct_all_symbol = symbol 
        
    elif float_ftd_pct_2021 > greatest_float_ftd_pct_2021:
        greatest_float_ftd_pct_2021 = float_ftd_pct_2021
        greatest_float_ftd_pct_2021_symbol = symbol 
        
    elif float_ftd_pct_ytd > greatest_float_ftd_pct_ytd:
        greatest_float_ftd_pct_ytd = float_ftd_pct_ytd
        greatest_float_ftd_pct_ytd_symbol = symbol

## End for loop 
    

## Outstanding Shares 
print('greatest_os_ftd_pct_all_symbol')
print(greatest_os_ftd_pct_all_symbol)
print(greatest_os_ftd_pct_all)
print()

print('greatest_os_ftd_pct_2021_symbol')
print(greatest_os_ftd_pct_2021_symbol)
print(greatest_os_ftd_pct_2021)
print()

print('greatest_os_ftd_pct_ytd_symbol')
print(greatest_os_ftd_pct_ytd_symbol)
print(greatest_os_ftd_pct_ytd)
print()

## Float 
print('greatest_float_ftd_pct_all_symbol')
print(greatest_float_ftd_pct_all_symbol)
print(greatest_float_ftd_pct_all)
print()

print('greatest_float_ftd_pct_2021_symbol')
print(greatest_float_ftd_pct_2021_symbol)
print(greatest_float_ftd_pct_2021)
print()

print('greatest_float_ftd_pct_ytd_symbol')
print(greatest_float_ftd_pct_ytd_symbol)
print(greatest_float_ftd_pct_ytd)
print()
```

```python
ftd_short_info_dict = {}

for symbol in datadict_success_list:
    import_path = '../FilesExport_Complete_DataDicts/'+symbol+'_data_dict.pkl'    
    data = load_obj(import_path)
    
    ## os_ftd_pct_all - all time count of FTDs for the dataset, relative to outstanding shares 
    os_ftd_pct_all = data[symbol]['ftd_stats']['os_ftd_pct_all']
    ## os_ftd_pct_2021 - pct of FTDs relative to outstanding shares, for Jan1st to Dec31st 2021
    os_ftd_pct_2021 = data[symbol]['ftd_stats']['os_ftd_pct_2021']
    ## os_ftd_pct_ytd - pct of FTDs relative to outstanding shares, for Dec1st 2020 to Dec31st 2021
    os_ftd_pct_ytd = data[symbol]['ftd_stats']['os_ftd_pct_ytd']
    
    ## float_ftd_pct_all - all time count of FTDs for the dataset, relative to Float shares 
    float_ftd_pct_all = data[symbol]['ftd_stats']['float_ftd_pct_all']
    ## float_ftd_pct_2021 - pct of FTDs relative to Float shares, for Jan1st to Dec31st 2021
    float_ftd_pct_2021 = data[symbol]['ftd_stats']['float_ftd_pct_2021']
    ## float_ftd_pct_ytd - pct of FTDs relative to Float shares, for Dec1st 2020 to Dec31st 2021
    float_ftd_pct_ytd = data[symbol]['ftd_stats']['float_ftd_pct_ytd'] 
    
    
    short_volume_NSDQpct_all_data = data[symbol]['ftd_stats']['short_volume_NSDQpct_all_data']
    short_volume_NSDQpct_ytd = data[symbol]['ftd_stats']['short_volume_NSDQpct_ytd']
    short_volume_FMPpct_all_data = data[symbol]['ftd_stats']['short_volume_FMPpct_all_data']
    short_volume_FMPpct_all_data = data[symbol]['ftd_stats']['short_volume_FMPpct_all_data']
    
    ftd_short_info_dict[symbol] = {
        'os_ftd_pct_all':os_ftd_pct_all,
        'os_ftd_pct_2021':os_ftd_pct_2021,
        'os_ftd_pct_ytd':os_ftd_pct_ytd,
        'float_ftd_pct_all':float_ftd_pct_all,
        'float_ftd_pct_2021':float_ftd_pct_2021,
        'float_ftd_pct_ytd':float_ftd_pct_ytd,
        'short_volume_NSDQpct_all_data':short_volume_NSDQpct_all_data,
        'short_volume_NSDQpct_ytd':short_volume_NSDQpct_ytd,
        'short_volume_FMPpct_all_data':short_volume_FMPpct_all_data,
        'short_volume_FMPpct_all_data':short_volume_FMPpct_all_data        
    }

    
    
len(ftd_short_info_dict)
```

```python
## Create list of symbols with greater than 5% of FTDs in the past year 
machine_learning_list = []
machine_learning_dict = {}


for symbol in ftd_short_info_dict:
    float_ftd_pct_ytd = ftd_short_info_dict[symbol]['float_ftd_pct_ytd']
    os_ftd_pct_ytd = ftd_short_info_dict[symbol]['os_ftd_pct_ytd']
    short_volume_NSDQpct_ytd = ftd_short_info_dict[symbol]['short_volume_NSDQpct_ytd']
    isETF = False
    
    if float_ftd_pct_ytd == 0:
        float_ftd_pct_ytd = os_ftd_pct_ytd
        isETF = True
        
    ## If 5% or more of the (float or outstanding shares) has failed to deliver in the past year, 
    ## and if the ytd SI metric represents more than 5% of the stock traded as short...
    if (float_ftd_pct_ytd >= 5.0) and (short_volume_NSDQpct_ytd>= 20.0):
        machine_learning_list.append(symbol)
        machine_learning_dict[symbol] = {
            'ytd_FTD_pct' : float_ftd_pct_ytd,
            'ytd_SI_pct'  : short_volume_NSDQpct_ytd,
            'isETF':isETF
        }
        
print(len(machine_learning_list))
print(len(machine_learning_dict))
```
